**What Alcohol Consumption Over Time (1977-2018) tells us -**
1977 had the highest Alcohol Consumption in gallons per capita, if we look at the bar graph we can see that there is a steady decline
with slight increases in specific years.  

**Lowest recorded Alcohol Consumption Over Time -**
From 1977 to 1995 there was a steady decline, and in 1995 gallons per capita hit its lowest recorded value.  
This could of happened because of changes with the government or programs that were put in place to bring alcohol consumption down in the U.S

**Increase from 1999 to 2018 -**
After the lowest recorded gallons per capita we can see that the consumption slowly increased from 1999 to 2018 with very slight decreases
in specific years.

```python
# Bar graph for total alcohol consumption over time
plt.figure(figsize=(12, 6))  # Adjust the figure size
plt.bar(alcohol_data['year'], alcohol_data['ethanol_all_drinks_gallons_per_capita'], color='skyblue')
plt.title('Alcohol Consumption Over Time (1977-2018)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Alcohol Consumption (gallons per capita)', fontsize=12)
plt.xticks (rotation=45)  
plt.yticks(fontsize=10)
plt.show()
```

**Downward Trend -** In this scatter plot we can see a downward trend from the highest recorded which was in 1977. 

**Stability -** From 1995 - 2001 we can see that this was a period of time when the alcohol consumption remained stable.

**Significant changes -** In 2011 there was a significant spike in alcohol consumption which set the tone for the years to come up until 2018.
After the spike we can see that the alcohol consumption still remained on that upward trend without any major decreases. 

```python
plt.figure(figsize=(12, 8))


plt.scatter(alcohol_data['year'], alcohol_data['ethanol_all_drinks_gallons_per_capita'])

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Alcohol Consumption (Gallons per Capita)')
plt.title('Alcohol Consumption Over Time (1977-2018)')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.jointplot(kind='reg', data=alcohol_data, 
              x='year', y='ethanol_all_drinks_gallons_per_capita')

plt.show()
```

**Top 10 states with Highest Average Alcohol Consumption -** This graph showed us which states had the highest average
Alcohol consumption from 1977-2018. 

```python
state_avg_consumption = alcohol_data.groupby('state')['ethanol_all_drinks_gallons_per_capita'].mean()

top_states = state_avg_consumption.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
top_states.plot(kind='bar', color='skyblue')

plt.xlabel('State')
plt.ylabel('Average Alcohol Consumption (Gallons per Capita)')
plt.title('Top 10 States with Highest Average Alcohol Consumption (1977-2018)')

plt.xticks(rotation=45)
plt.show()
```
