#### OlS Regression Tests - Total Alcohol Consumption, Beer and Shots
Focusing on OLS regressions for total alcohol consumption, beer, and shots will help us answer our original research question, as we can analyze the trends between these variables. Since beer and shots contribute the most to the average total number of drinks, comparing their trends will provide insight into whether alcohol consumption is increasing or decreasing in the U.S. between 1977-2018. By examining these regressions separately, we can determine if their patterns align or differ, offering a clearer picture of alcohol consumption trends.

**OLS Regression for Total alcohol consumption overtime**
  The OLS Regression results will help us to better understand whether alcohol consumption has increased or decreased in the U.S. The test shows a 
negative, with consumption decreasing by -0.0082 gallons per year on average.  There are other factors in the data that led to the slow decline. We will look at other variables OLS regressions to see what alcohol types held the impact of alcohol consumption to slowly decrease. Since beer was the highest contributor of total number of drinks total we will look at it's OLS Regression results next.  


```python
fit = smf.ols('ethanol_all_drinks_gallons_per_capita ~ year', data=alcohol_data).fit()

fit.summary()
```

**OLS Regression for number of beers** Beer was the #1 contributor to the average of total drinks consumed.  
We want to look at it's OLS Regression results to see how big of an impact it had to the overall decline for total alcohol consumption overtime. 
This Regression also shows a negative of -1.4612 for the year coef which indicates a significant decrease in beer consumption. We will analyze one more
OLS regression for numbers of shots because this was the second highest contributor to the average number of drinks total.


```python
fit = smf.ols('number_of_beers ~ year', data=alcohol_data).fit()

# Print the summary of the regression
fit.summary()
```

**OLS Regression for number of shots of liquor** Liquor was the #2 contributor to the average of total drinks consumed. The coef indicated
a negative of -1.0558 which helps us understand how alcohol consumption is declining since shots are the second highest average of drinks consumed.  

```python
fit = smf.ols('number_of_shots_liquor ~ year', data=alcohol_data).fit()

# Print the summary of the regression
fit.summary()
```

#### Bar graph for Total Alcohol consumption over time
This bar graph illustrates the overall trend of total alcohol consumption (gallons per capita) in the U.S. from 1977 to 2018. By observing the fluctuations
in the graph we can analyze whether alcohol consumption is generally increasing or decreasing.  This graph supports our research question by
giving us an overview of trends which helps us understand how alcohol consumption has shifted from 1977-2018. 

```python
plt.figure(figsize=(12, 6))  
plt.bar(alcohol_data['year'], alcohol_data['ethanol_all_drinks_gallons_per_capita'], color='skyblue')
plt.title('Alcohol Consumption Over Time (1977-2018)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Alcohol Consumption (gallons per capita)', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```
